```python
import duckdb

# File paths (update as needed)
sabina_file = 'sabina_reduced.parquet'  # Reduced file with 2 records
climate_file = 'climate_merged_final.parquet'

# Connect to DuckDB
con = duckdb.connect()

# Perform INNER JOIN on first two characters of Latitude + OBSERVATION DATE
query = f"""
    SELECT sab.*, cli.Latitude, cli.Longitude, cli.Elevation
    FROM (
        SELECT *, LEFT(CAST(LATITUDE AS VARCHAR), 4) AS lat_prefix
        FROM read_parquet('{sabina_file}')
        LIMIT 1
    ) AS sab
    INNER JOIN (
        SELECT *, LEFT(CAST(Latitude AS VARCHAR), 4) AS lat_prefix
        FROM read_parquet('{climate_file}')
    ) AS cli
    ON sab."OBSERVATION DATE" = cli.Date
    AND sab.lat_prefix = cli.lat_prefix
"""

# Execute the query and store result in a Pandas DataFrame
df_merged = con.execute(query).df()

# Display the merged records
print(df_merged)
```

```python
import duckdb

# File paths (update as needed)
sabina_file = 'sabina_reduced.parquet'  # Reduced file with 2 records
climate_file = 'climate_merged_final.parquet'

# Connect to DuckDB
con = duckdb.connect()

# Query to join Sabina with the closest Climate Latitude & Longitude
query = f"""
    WITH RankedMatches AS (
        SELECT
            sab.*,
            cli.Latitude AS Climate_Latitude,
            cli.Longitude AS Climate_Longitude,
            cli.Elevation,
            ABS(sab.LATITUDE - cli.Latitude) + ABS(sab.LONGITUDE - cli.Longitude) AS distance,
            ROW_NUMBER() OVER (
                PARTITION BY sab.LATITUDE, sab.LONGITUDE, sab."OBSERVATION DATE"
                ORDER BY ABS(sab.LATITUDE - cli.Latitude) + ABS(sab.LONGITUDE - cli.Longitude) ASC
            ) AS rank
        FROM read_parquet('{sabina_file}') AS sab
        LEFT JOIN read_parquet('{climate_file}') AS cli
        ON sab."OBSERVATION DATE" = cli.Date
    )
    SELECT * FROM RankedMatches WHERE rank = 1
"""

# Execute the query and store result in a Pandas DataFrame
df_merged = con.execute(query).df()

# Display the merged records
print(df_merged)
```

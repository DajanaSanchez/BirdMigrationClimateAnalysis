```python
#  Round to 6 decimal places (approximately 0.11 meters precision)
df_bird["Latitude"] = df_bird["Latitude"].round(6)
df_bird["Longitude"] = df_bird["Longitude"].round(6)

df_MODIS["Latitude"] = df_MODIS["Latitude"].round(6)
df_MODIS["Longitude"] = df_MODIS["Longitude"].round(6)

df_climate["Latitude"] = df_climate["Latitude"].round(6)
df_climate["Longitude"] = df_climate["Longitude"].round(6)
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point


#  Convert Land Data to GeoDataFrame
land_gdf = gpd.GeoDataFrame(
    df_MODIS,
    geometry=gpd.points_from_xy(df_MODIS.Longitude, df_MODIS.Latitude),
    crs="EPSG:4326"  # WGS84 Coordinate Reference System
)


#  Convert Bird Data to GeoDataFrame
bird_gdf = gpd.GeoDataFrame(
    df_bird,
    geometry=gpd.points_from_xy(df_bird.Longitude, df_bird.Latitude),
    crs="EPSG:4326"
)
```

```python
# Round coordinates again after conversion
bird_gdf["Latitude"] = bird_gdf["Latitude"].round(6)
bird_gdf["Longitude"] = bird_gdf["Longitude"].round(6)
land_gdf["Latitude"] = land_gdf["Latitude"].round(6)
land_gdf["Longitude"] = land_gdf["Longitude"].round(6)
```

```python
# Convert both datasets to EPSG:3857 (Latitude/Longitude)
#  Reduce Land Data to Only Regions Close to Bird Observations
#  Define a larger buffer for Climate Data (Ensures no missing matches)
buffer_size = 1.5  # ~100km search radius

#  Reduce datasets to only relevant regions
bird_bounds = bird_gdf.total_bounds  # (minx, miny, maxx, maxy)

land_gdf = land_gdf.cx[bird_bounds[0] - buffer_size : bird_bounds[2] + buffer_size,
                        bird_bounds[1] - buffer_size : bird_bounds[3] + buffer_size]

print(f" Reduced land dataset size: {len(land_gdf)} records.")
print(f" Reduced climate dataset size: {len(climate_gdf)} records.")

#  Convert All to EPSG:3857 for Accurate Distance Calculations
bird_gdf = bird_gdf.to_crs("EPSG:3857")
land_gdf = land_gdf.to_crs("EPSG:3857")
```

```python
import os
import gc

#  Define Paths in Google Drive
base_dir = "/content/drive/MyDrive/bird_data_processing"
#  Ensure directory exists
os.makedirs(base_dir, exist_ok=True)

#  Reduce Memory Usage by Converting to float32
for df in [bird_gdf, land_gdf]:
    df["Latitude"] = df["Latitude"].astype("float32")
    df["Longitude"] = df["Longitude"].astype("float32")

#  Save to Parquet for Efficient Processing
print(" Saving datasets to Parquet for efficient chunked processing...")
bird_parquet = os.path.join(base_dir, "bird_data.parquet")
land_parquet = os.path.join(base_dir, "land_data.parquet")

bird_gdf.to_parquet(bird_parquet, index=False)
land_gdf.to_parquet(land_parquet, index=False)

#  Free Memory
del bird_gdf, land_gdf
gc.collect()
```

```python
import os
from dask.diagnostics import ProgressBar

#  Define Chunk Parameters
output_dir = os.path.join(base_dir, "processed_chunks")
os.makedirs(output_dir, exist_ok=True)
chunk_size = 250000  #  Adjust chunk size for efficiency

existing_chunks = set(os.listdir(output_dir))

print(" Loading datasets from Parquet...")
bird_ddf = pd.read_parquet(bird_parquet, columns=["Latitude", "Longitude", "Date"])
land_ddf = pd.read_parquet(land_parquet, columns=["Latitude", "Longitude", "Land_Cover_Type"])

print(" Starting optimized chunked processing...")
with ProgressBar():
    for i in range(0, len(bird_ddf), chunk_size):
        chunk_file = f"bird_land_chunk_{i}.parquet"
        chunk_path = os.path.join(output_dir, chunk_file)

        #  Skip if already processed
        if chunk_file in existing_chunks:
            print(f" Skipping {chunk_file}")
            continue

        print(f" Processing rows {i} to {i + chunk_size}...")

        #  Load bird chunk
        bird_ddf = pd.read_parquet(bird_parquet, columns=["Latitude", "Longitude", "Date"])  # Load full data once
        bird_chunk = bird_ddf.iloc[i : i + chunk_size]  # Select the current chunk

        #  Convert to GeoDataFrame
        bird_gdf = gpd.GeoDataFrame(
            bird_chunk,
            geometry=gpd.points_from_xy(bird_chunk.Longitude, bird_chunk.Latitude),
            crs="EPSG:3857"
        )

        #  Reduce land and climate datasets to nearby regions
        buffer_size = 1.5  # ~150km search radius
        land_gdf = pd.read_parquet(land_parquet, columns=["Latitude", "Longitude", "Land_Cover_Type"])
        climate_gdf = pd.read_parquet(climate_parquet, columns=["Latitude", "Longitude", "Temp_celsius", "Precip_mm"])

        land_gdf = land_gdf[
            (land_gdf["Latitude"] >= bird_gdf.total_bounds[1] - buffer_size) &
            (land_gdf["Latitude"] <= bird_gdf.total_bounds[3] + buffer_size) &
            (land_gdf["Longitude"] >= bird_gdf.total_bounds[0] - buffer_size) &
            (land_gdf["Longitude"] <= bird_gdf.total_bounds[2] + buffer_size)
        ]

        #  Convert to GeoDataFrame for Spatial Join
        land_gdf = gpd.GeoDataFrame(land_gdf, geometry=gpd.points_from_xy(land_gdf.Longitude, land_gdf.Latitude), crs="EPSG:3857")

        #  Perform Spatial Joins with Nearest Match (Adding max_distance)
        bird_land_chunk = gpd.sjoin_nearest(
            bird_gdf, land_gdf, how="left", distance_col="land_distance", max_distance=50_000
        )


        #  Compute Year Differences
        bird_land_chunk["Year_Diff_Land"] = abs(bird_land_chunk["Date"] - bird_land_chunk["Year_right"])
        bird_land_chunk["Year_Diff_Climate"] = abs(bird_land_chunk["Date"] - bird_land_chunk["Year_right_Climate"])

        #  Keep Only the Closest Match
        bird_land_chunk = bird_land_chunk.sort_values(
            by=["Latitude", "Longitude", "Date", "Year_Diff_Land", "land_distance",
        ).drop_duplicates(subset=["Latitude", "Longitude", "Date"], keep="first")

        #  Convert Back to Geographic CRS **Only for Saving**
        bird_land_chunk = bird_land_chunk.to_crs("EPSG:4326")

        #  Save Each Chunk Directly to Google Drive
        bird_land_chunk.to_parquet(chunk_path, index=False)

        print(f" Processed and saved {len(bird_land_chunk)} rows to {chunk_path}.")

        #  Free Memory
        del bird_gdf, land_gdf, climate_gdf, bird_land_chunk
        gc.collect()
```

```python
import os
import pandas as pd
import gc
from fastparquet import write

#  Define Paths
chunk_dir = "/content/drive/MyDrive/bird_land_chunks"
final_output_file = "/content/drive/MyDrive/bird_land_spatial_updated.parquet"

#  Check if `final_output_file` is a directory and delete it
if os.path.exists(final_output_file) and os.path.isdir(final_output_file):
    print(f" Found a directory named {final_output_file}. Deleting it...")
    os.rmdir(final_output_file)  #  Delete the directory to allow file writing

#  Get List of All Chunked Files
all_chunks = sorted([os.path.join(chunk_dir, f) for f in os.listdir(chunk_dir) if f.endswith(".parquet")])

print(f" Found {len(all_chunks)} chunks to merge.")

if len(all_chunks) == 0:
    print(" No chunk files found. Ensure processing is complete before merging.")
else:
    #  Write the First Chunk as the Base File
    first_file = all_chunks.pop(0)  # Take first chunk
    print(f" Initializing final file with {first_file}...")

    df = pd.read_parquet(first_file)  #  Read first chunk
    write(final_output_file, df, compression="snappy")  #  Write first chunk to Parquet

    #  Append Remaining Chunks One by One
    for chunk_file in all_chunks:
        print(f" Merging {chunk_file}...")

        df_chunk = pd.read_parquet(chunk_file)  #  Read chunk

        #  Append chunk without loading everything into memory
        write(final_output_file, df_chunk, append=True, compression="snappy")  #  Append new chunk

        #  Free Memory
        del df_chunk
        gc.collect()

        print(f" Merged {chunk_file} successfully.")

    print(f" Final dataset saved at: {final_output_file}")
```

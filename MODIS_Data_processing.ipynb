```python
import pandas as pd

# Path to the merged dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/Capstone/Processed_Data/merged_dataset.csv'

# Load the dataset
merged_dataset = pd.read_csv(file_path)

# Display the first few rows to confirm
merged_dataset.head()
```

```python
import matplotlib.pyplot as plt

# Group the data by Year and calculate the total occurrences of each land cover type
land_cover_trends = merged_dataset.groupby(['Year', 'Land_Cover_Description']).size().unstack().fillna(0)

# Plot the trends of different land cover types over time
land_cover_trends.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Trends of Land Cover Types Over Time')
plt.xlabel('Year')
plt.ylabel('Land Cover Count')
plt.legend(title='Land Cover Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
